```python
import urllib.request
from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/Florida"
page = urllib.request.urlopen(url)
soup = BeautifulSoup(page, "lxml")
print(soup.prettify())
soup.title.string

```

```python
table = soup.find('table', class_="toccolours")
table
```

```python
for row in table.find_all('tr'):
    cells = row.find_all('td')
    print(len(cells))
    for column in cells:
        print(column)
```

```python
for row in table.find_all('tr'):
    cells = row.find_all('td')
    print(len(cells))
    for column in cells:
        print(column)
```

```python
A = []
B = []
C = []


for row in table.find_all('tr'):
    cells = row.find_all('td')
    if len(cells) == 4:
        A.append(cells[0].find(text=True))
        B.append(cells[1].find(text=True))
        C.append(cells[3].find(text=True))
```

```python
import pandas as pd

df = pd.DataFrame(A, columns=['Year'])
df['Population'] = B
df['%'] = C
df
```

```python

from matplotlib import pyplot as plt



plt.figure(figsize=(15,10))
plt.plot(B, linestyle='-.', label= "Population")
plt.xticks(range(0,len(B)),A)
plt.title("Population", fontsize=20)
plt.legend(loc='upper right', fontsize=14)
plt.xlabel("Population")
plt.show()
```
